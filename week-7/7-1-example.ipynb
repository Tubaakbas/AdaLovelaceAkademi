# Linear ve Non-Liear Regresyon

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pylab as pylab
%matplotlib inline
```

```python
x= np.arange(-5.0, 5.0, 0.1)

#Linear regresyon formülü
y = 3 + 2*(x)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
```

```python
plt.figure(figsize=(12,8))
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
df = pd.read_csv("Data/china_gdp.csv")
```

```python
df.head(10)
```

```python
x_data, y_data=(df["Year"].values, df["Value"].values)
plt.plot(x_data, y_data, 'go')
plt.xlabel("Year")
plt.ylabel("GDB")
plt.show()
```

```python
#veriye Exponential eğri uydurmak için gerekli olan kütüphane
from scipy.optimize import curve_fit 
```

```python
# Sigmoid model fonksiyonu
def sigmoid_function(x, a, b):
    return 1 / (1 + np.exp(-a * (x - b)))
```

params: exponential_function'ın parametrelerini içeren bir tuple 

covariance: parametre tahminlerinin doğruluğunu gösteren bir kovaryans matrisidir. Kovaryans matrisi, parametre tahminlerinin belirsizliklerini ve bu tahminler arasındaki ilişkileri ifade eder.

```python
# Başlangıç parametrelerini manuel olarak belirle
initial_params = [1.0, 2000.0]  # Bu değerleri ihtiyaca göre ayarlayabilirsiniz

# Sigmoid fonksiyonunu optimize et
params, covariance = curve_fit(sigmoid_function, x_data, y_data, p0=initial_params, maxfev=100000)
```

```python
print(params)
```

```python
# Optimize edilmiş parametreleri al
a, b = params

# Tahmin edilen y değerlerini hesapla
predict_y = sigmoid_function(x_data, a, b)
```

```python
# Veriyi ve sigmoid fonksiyonunun tahminini görselleştir
plt.scatter(x_data, y_data, color='blue', label='Veri')
plt.plot(x_data, predict_y, color='red', label='Sigmoid Model')
plt.xlabel("Year (Yıl)")
plt.ylabel("GDP (Mal ve Hizmetlerin Toplamı)")
plt.legend()#labeldaki değerleri yazar
plt.show()
```
