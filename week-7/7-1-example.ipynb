# Linear ve Non-Liear Regresyon

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pylab as pylab
%matplotlib inline
```

```python
x= np.arange(-5.0, 5.0, 0.1)

#Linear regresyon formülü
y = 3 + 2*(x)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
```

```python
plt.figure(figsize=(12,8))
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
df = pd.read_csv("Data/china_gdp.csv")
```

```python
df.head(10)
```

```python
x_data, y_data=(df["Year"].values, df["Value"].values)
plt.plot(x_data, y_data, 'go')
plt.xlabel("Year")
plt.ylabel("GDB")
plt.show()
```

```python
#veriye Exponential eğri uydurmak için gerekli olan kütüphane
from scipy.optimize import curve_fit 
```

```python
# Exponential model fonksiyonu
def exponential_function(x, a, b):
    return a * np.exp(b * x)

def sigmoid_function(x, m, n):
    return 1 / (1 + np.exp(-m * (x - n)))
```

params: exponential_function'ın parametrelerini içeren bir tuple 

covariance: Bu, parametre tahminlerinin doğruluğunu gösteren bir kovaryans matrisidir. Kovaryans matrisi, parametre tahminlerinin belirsizliklerini ve bu tahminler arasındaki ilişkileri ifade eder.

```python
params, covariance= curve_fit(exponential_function, x_data, y_data)
params_, covariance_ = curve_fit(sigmoid_function, x_data, y_data)
```

```python
a, b=params #params'dan gelen değerler
m, n=params_
```

```python
# Modelin tahmin ettiği değerleri hesapla
print("Exponential")
predict_y_exponential= exponential_function(x_data, a, b)
print(predict_y_exponential)

print("Sigmoid")
predict_y_sigmoid=sigmoid_function(x_data, m, n)
print(predict_y_sigmoid)
```

```python
#Görselleştirme - Exponential
plt.plot(x_data, y_data,  'bo')
plt.plot(x_data, predict_y_exponential, 'r') 
plt.xlabel("Year (Yıl)")
plt.ylabel("GDP (Mal ve Hizmetlerin Toplamı)")
plt.show()
```

```python
#Görselleştirme - Sigmoid
plt.plot(x_data, y_data,  'bo')
plt.plot(x_data, predict_y_sigmoid, 'r') 
plt.xlabel("Year (Yıl)")
plt.ylabel("GDP (Mal ve Hizmetlerin Toplamı)")
plt.show()
```
