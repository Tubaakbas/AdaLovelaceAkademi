```python
def article_search_information(df): 
    for article_search in response['response']['docs']:
        #yukarÄ±da yapÄ±lan iÅlemler dÃ¼zenli Åekilde yazÄ±ldÄ±
        type_of_material=article_search['type_of_material']
        word_count=article_search['word_count']
        abstract=article_search['abstract']

        keywords_name=article_search['keywords'][0]['name']
        keywords_value=article_search['keywords'][0]['value']
        
        #doÄruluÄuna bakabilmek iÃ§in ekrana yazdÄ±rÄ±ldÄ±   
        print(abstract)
        print(type_of_material)
        print(word_count)
        print(keywords_name)
        print(keywords_value)
            
        print("\n") #her bir dÃ¶ngÃ¼den sonra boÅluk bÄ±rakma
            
        #save in pandas df
        df = df._append({
            "abstract": abstract,
            "type_of_material": type_of_material,
            "word_count": word_count,
            "keywords_name": keywords_name,
            "keywords_value": keywords_value
            }, 
        ignore_index = True)
    return df
```

```python
df = pd.DataFrame(columns=["abstract", "type_of_material", "word_count", "keywords_name", "keywords_value"]) #kolon isimleri eklendi
ny_times_data = article_search_information(df)
```
