```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
iris=load_iris()#numpy array olduğu için
#dataframe e çevriliyor ve kolonları yazıyoruz
iris_df=pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
```

Korelasyon katsayısı negatif ise iki değişken arasında ters ilişki vardır,pozitif ise doğrusal ilişki vardır.

```python
iris_df.corr()
```

```python
#Isı Haritasını kolerasyon kat sayılarını daha iyi okuyabilmek için kullanabiliriz.
corr=iris_df.corr()
sns.heatmap(corr,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values);
```
