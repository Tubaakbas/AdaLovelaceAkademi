```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches
import seaborn as sns
import numpy as np  
import pandas as pd
```

```python
df= pd.DataFrame({
    
    
    'Group': ['A', 'B', 'C', 'D', 'E' ],
    'Value': [1, 5, 4, 3, 9]
})
```

```python
plt.barh(y=df.Group, width=df.Value)
```

```python
df=df.sort_values(by=['Value'])
my_plot= plt.barh(
    y=df.Group, 
    width=df.Value, 
    #color=['yellow', 'pink', 'grey'],
    color=mcolors.TABLEAU_COLORS, 
    #https://matplotlib.org/stable/gallery/color/named_colors.html
    edgecolor='black',#kenar rengi
    linewidth=2, #kenar kalınlığı    
    )
plt.bar_label(container=my_plot, padding=3)
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.barplot(
    x= "total_bill",
    y="day",
    data=tips
    
)
```

```python
sns.set(style="darkgrid")
sns.barplot(
    x="day",
    y="total_bill",
    data=tips,
    color="pink",
    hue="sex"
)
plt.figure(figsize=(18, 8))
```

```python
tips.head()
```

```python
#data_frame haline getirmek için reset_index kullanılır
total=tips.groupby('day')['total_bill'].sum().reset_index() #her gün için toplam ödeme
```

```python
total
```

```python
bar1=sns.barplot(x="day", y="total_bill", data=total, color='darkblue')
```

```python
smoker=tips[tips.smoker=='Yes']
```

```python
bar2=sns.barplot(
    x='day',
    y='total_bill',
    data=smoker,
    estimator=sum,
    errorbar=None,
    color='lightblue')
```

```python

bar1=sns.barplot(
    x="day", 
    y="total_bill", 
    data=total, 
    color='darkblue')

bar2=sns.barplot(
    x='day',
    y='total_bill',
    data=smoker,
    estimator=sum,
    errorbar=None,
    color='lightblue')

top_bar=mpatches.Patch(color='darkblue', label='non-smoker')
bottom_bar=mpatches.Patch(color='lightblue', label='smoker=Yes')
plt.legend(handles=[top_bar, bottom_bar])
plt.show()
```
